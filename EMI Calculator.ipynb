```python
#...
# Rashi Saluja
# salujarashi333@gmail.com
#...


# asking the user for vehicle's purchase price, down payment and sales tax
purchase_price = float(input("Enter the vehicle's purchase price: "))
down_payment = float(input("Enter the down payment for the vehicle: "))
sales_tax = float(input("Enter the sales tax on the transaction (for 8.25% tax, enter 8.25): "))


# checking if the user has a vehicle to trade in
choice = input("Do you have a vehicle to trade in? (y/n): ")
if choice.lower() == 'y':
    trade_in_value = float(input("Enter the trade-in value of your current vehicle: "))
    amount_owed = float(input("Enter the amount owed on your current vehicle: "))
else:
    trade_in_value = 0
    amount_owed = 0

    
# calculating and printing overall down payment and overall loan amount
total_down_payment = down_payment + (trade_in_value - amount_owed)
total_loan_amount = (purchase_price - total_down_payment) * (1 + (sales_tax / 100))   
print("Your overall down payment is $" + str(total_down_payment))
print("Your overall loan amount is $" + str(total_loan_amount))

print("**********************************")


# displaying the menu and finding the number of months and interest rate
percent_down = (total_down_payment/purchase_price)*100
print("Enter the length of your loan...\n\t1: 3 years\n\t2: 4 years\n\t3: 5 years\n\t4: 6 years")
option = input("Select an option: ")
if option == '1':
    year = 3
    if percent_down<20:
        interest_rate = 4
    else:
        interest_rate = 3.7   
elif option == '2':
    year = 4
    if percent_down<20:
        interest_rate = 4.33
    else:
        interest_rate = 3.8
elif option == '3':
    year = 5
    if percent_down<20:
        interest_rate = 4.66
    else:
        interest_rate = 3.9
elif option == '4':
    year = 6
    if percent_down<20:
        interest_rate = 5
    else:
        interest_rate = 4
else:
    year = 5
    if percent_down<20:
        interest_rate = 4.66
    else:
        interest_rate = 3.9
print("You selected " + str(year) + " year with a total of " + str(year*12) + " monthly payments.")


print("**********************************")

print("With "+str(percent_down)+"% down and a "+str(year)+" year loan, we can offer you an interest rate of "+str(interest_rate)+"%.")

print("**********************************")


# monthly payment calculation
monthly_interest = interest_rate/1200
b = 1-((1+monthly_interest)**-(year*12))
a = monthly_interest/b
monthly_payment = total_loan_amount*a
print("Your estimated monthly payment would be $"+str(monthly_payment)+" a month.")




```
