```python


# Introduction for the user

print("INTRODUCTION\n")
print("This is a Python Powered Unit Converter. You need to provide")
print("a quantity to convert, the unit to convert and the unit to be")
print("converted in.\nThe units are divided into three categories: ")
print("1. Distances: 'ft', 'cm', 'mm', 'mi', 'm', 'yd', 'km', 'in'")
print("2. Weights: 'lb', 'mg','kg', 'oz', 'g'")
print("3. Volumes: 'floz', 'qt', 'cup', 'mL', 'L', 'gal', 'pint'\n")
print("Make sure that the unit you are converting must match the")
print("category with the unit you are converting into.")
print("For example, kg converting into m should not happen.\n")





import ipywidgets as widgets

value = widgets.FloatText(
    value=10.0,
    description='Convert: ',
    disabled=False
)

convertFrom = widgets.Dropdown(
    options=['ft', 'cm', 'mm', 'mi', 'm', 'yd', 'km', 'in', #distances
             'lb', 'mg','kg', 'oz', 'g', #weights
             'floz', 'qt', 'cup', 'mL', 'L', 'gal', 'pint'],#volumes
    value='mi',
    description='',
    disabled=False,
)

convertTo = widgets.Dropdown(
    options=['ft', 'cm', 'mm', 'mi', 'm', 'yd', 'km', 'in', #distances
             'lb', 'mg','kg', 'oz', 'g', #weights
             'floz', 'qt', 'cup', 'mL', 'L', 'gal', 'pint'],#volumes
    value='m',
    description='to ',
    disabled=False,
)



items = [value, convertFrom, convertTo]
left_box = widgets.VBox([items[0], items[2]])
right_box = widgets.VBox([items[1]])
widgets.HBox([left_box, right_box])


```

```python


# function checkInput() which accepts three dictionaries originally defined in main
# and returns integer according to the input of the user
def checkInput(distance, weight , volume):
    if convertFrom.value in distance:
        if convertTo.value not in distance:
            print("The units do not match the same category. Please refer the introduction.")
        else:
            return 1
           
    elif convertFrom.value in weight:
        if convertTo.value not in weight:
            print("The units do not match the same category. Please refer the introduction.")
        else:
            return 2
            
    else:
        if convertTo.value not in volume:
            print("The units do not match the same category. Please refer the introduction.")
        else:
            return 3
            
    
# function calc() which accepts a dictionary and a float value, does the calculation 
# and displays the output
def calc(temp_dict , val):
    for i in temp_dict:
        if i == convertFrom.value:
            a = val * temp_dict[i]
            break
    for i in temp_dict:
        if i == convertTo.value:
            b = a / temp_dict[i]
            break
    print(str(val)+" "+convertFrom.value+" is equal to "+ str(b)+" "+convertTo.value)
    
        
    

    
def main():
    
    # Dictionaries to store all the units as keys and a constant as their values.
    # This constant helps the unit to convert into base unit and then the user desired unit.
    # For diastance the base unit is m, similarly for weight it is kg and for volume it is L.
    distance_dict = {'ft':0.3048, 'cm':0.01, 'mm':0.001, 'mi':1609.34, 'm':1, 'yd':0.9144, 'km':1000, 'in':0.0254}
    weight_dict = {'lb':0.453592, 'mg':0.000001,'kg':1, 'oz':0.0283495, 'g':0.001}
    volume_dict = {'floz':0.0295735, 'qt':0.946353, 'cup':0.236588, 'mL':0.001, 'L':1, 'gal':3.78541, 'pint':0.473176}
    
    
    check = checkInput(distance_dict, weight_dict, volume_dict)
    
    # calculation for distance
    if check == 1:
        print("Let's do the calculation for distance \n")
        calc(distance_dict, value.value)
    
    # calculation for weight
    elif check == 2:
        print("Let's do the calculation for weight \n")
        calc(weight_dict, value.value)
    
    # calculation for volume
    elif check == 3:
        print("Let's do the calculation for volume \n")
        calc(volume_dict, value.value)
    
    # if the units do not match the category
    else:
        print("\nPlease make the changes accordingly and run the cell again.")

        
# calling main() function
main()
```
