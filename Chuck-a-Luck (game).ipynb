```python

# importing random module 
import random


# roll function for generating dice rolls
def roll():
    list1 = []
    for i in range(3):
        a = random.randrange(5)+1
        list1.append(a)
    return list1



# computeBetResult function for returning the amount won or lost
def computeBetResult(list1, bet_amount, guess):
    count = 0
    for i in list1:
        if i == int(guess):
            count += 1
    print("You matched "+str(count)+" dice!")
    if count == 0:
        return -int(bet_amount) 
    elif count == 1:
        return int(bet_amount)
    elif count == 2:
        return int(bet_amount)*3
    elif count == 3:
        return int(bet_amount)*10

    
# main function    
def main():
    user_money = 100
    print("STEP RIGHT UP AND PLAY SOME CHUCK-A-LUCK !\nYou have $100\n")
    choice = 'y'
    while choice.upper()== 'Y':
        
        
        
        # asking bet amount and checking the user input
        error = 1
        while error == 1:
            amount = input("Enter a bet amount: $")
            if not amount.isdigit():
                print("Invalid bet. Please enter an amount.")
            elif not (int(amount) > 0 and int(amount) <= int(user_money)):
                print("Invalid bet. Please enter an amount from $1 to $"+str(user_money))
            else:
                error = 0
        bet_amount = int(amount)
        
        
        
        # asking the number to bet on and checking the user input
        error1 = 1
        while error1 == 1:
            guess = input("What number do you want to bet on (1-6)? ")
            if not guess.isdigit():
                print("Invalid. Must be between 1 and 6.")
            elif not (int(guess)>=1 and int(guess)<=6):
                print("Invalid number. Must be between 1 and 6.")
            else:
                error1 = 0
        print("You bet $"+str(bet_amount)+" on "+str(guess))
        
        
        # calling roll function and printing the values
        l = roll()
        print("You rolled: ")
        for i in range(3):
            if i == 2:
                print(l[i])
            else:
                print(str(l[i]) , end = ",")
        
        
        
        # calling computeBetResult function and updating user's balance
        result = computeBetResult(l, bet_amount, guess)
        #print(result)
        user_money += result
        if result < 0:
            print("You lost your bet :(\nYou now have $"+str(user_money))
        else:
            print("YOU WIN $"+str(result)+"\nYou now have $"+str(user_money))
        
        
        # if the user runs out of money the game ends
        if user_money <= 0:
            break
        
        
        # asking for user's choice and checking the user input
        error2 = 1
        while error2 == 1:
            choice = input("\nWould you like to play again (y/n)? ")
            if choice != 'n' and choice != 'y':
                print("Invalid. Please enter y/n")
            else:
                error2 = 0
    
    
    # ending the game
    print("\nYou ended the game with $"+str(user_money))

    
# calling main function    
main()
```
