```python


Substitution_cypher = "EHTGDBWIUQRXLMVFSJCPYZKAON"
option = '1'

while option == '1' or option == '2':
    option = input("Select an option:\n1. Encrypt a message\n2. Decrypt a message\n?. Quit\n")
    
    # Encryption
    if option == '1':
        new = ""
        msg = input("Enter a message to encrypt : ")
        for i in range(0,len(msg)):
            
            # Encryption of upper case
            if ord(msg[i])>=65 and ord(msg[i])<=90:
                temp = ord(msg[i])-65
                a = Substitution_cypher[temp]
                
            # Encryption of lower case    
            elif ord(msg[i])>=97 and ord(msg[i])<=122:
                temp = ord(msg[i])-97
                a = Substitution_cypher[temp].lower()
                
            # Encryption of characters other than letters
            else:
                a = msg[i]
                
            # print(msg[i]+"\t"+str(ord(msg[i]))+"\t"+str(temp)+"\t"+Substitution_cypher[temp])
            
            # concatenation of secret message into a new string
            new += a
        print("The secret message is: "+new)   
        
    # Decryption
    elif option == '2':
        new = ""
        msg = input("Enter a message to decrypt : ")
        for i in range(0,len(msg)):
            
            # Decryption of upper case
            if ord(msg[i])>=65 and ord(msg[i])<=90:
                for k in range(0,len(Substitution_cypher)):
                    if msg[i] == Substitution_cypher[k]:
                        temp = k
                        temp1 = temp+65
                        a = chr(temp1)
                        
                        # print(msg[i]+"\t"+str(ord(msg[i]))+"\t"+str(k)+"\t"+str(temp1)+"\t"+a)
                        
            # Decryption of lower case
            elif ord(msg[i])>=97 and ord(msg[i])<=122:
                for k in range(0,len(Substitution_cypher)):
                    if msg[i].upper() == Substitution_cypher[k]:
                        temp = k
                        temp1 = temp+97
                        a = chr(temp1)
                        
            # Decryption of characters other than letters
            else:
                a = msg[i]
                
            # concatenation of secret message into a new string
            new +=a
        print("The secret message is: "+new) 
        
    # Quit
    else:
        print("Goodbye!")
    
```
